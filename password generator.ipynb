```python
import tkinter as tk
import random
import string

a = tk.Tk()
a.title("Password Generator")
a.config(bg='gray')

def generate_password():
  length = int(entry_length.get())
  characters = ""
  if check_uppercase.get():
    characters += string.ascii_uppercase
  if check_lowercase.get():
    characters += string.ascii_lowercase
  if check_digits.get():
    characters += string.digits
  if check_symbols.get():
    characters += string.punctuation

  password  = ''.join (random.choice(characters) for _ in range(length))
  entry_password.delete(0, tk.END)
  entry_password.insert(0, password)

# Label and entry for password length
label_length = tk.Label(a, text="Password Length:",font=('arial',10,'bold'),bg='magenta',fg='black')
label_length.pack(pady=5)
entry_length = tk.Entry(a, width=15,font=('arial',10,'bold'),bg='azure',fg='black')
entry_length.insert(0, 12)
entry_length.pack()

check_uppercase = tk.IntVar(value=1)
checkbox_uppercase = tk.Checkbutton(a, text="Uppercase Letters", variable=check_uppercase,font=('arial',10,'bold'),bg='turquoise',fg='black')
checkbox_uppercase.pack()

check_lowercase = tk.IntVar(value=1)
checkbox_lowercase = tk.Checkbutton(a, text="Lowercase Letters", variable=check_lowercase,font=('arial',10,'bold'),bg="turquoise",fg='black')
checkbox_lowercase.pack()

check_digits = tk.IntVar(value=1)
checkbox_digits = tk.Checkbutton(a, text="digits", variable=check_digits,font=('arial',10,'bold'),bg='turquoise',fg='black')
checkbox_digits.pack()

check_symbols = tk.IntVar(value=0)
checkbox_symbols = tk.Checkbutton(a, text="SYMBOLS", variable=check_symbols,font=('arial',10,'bold'),bg='turquoise',fg='black')
checkbox_symbols.pack()

label_password = tk.Label(a, text="Generated Password:",font=('arial',10,'bold'),bg='turquoise',fg='black')
label_password.pack(pady=5)
entry_password = tk.Entry(a, width=50,font=('arial',12,'bold'),bg='azure',fg='black')
entry_password.pack()

btn_generate = tk.Button(a, text="Generate Password", command=generate_password,font=('arial',10,'bold'),bg='pink',fg='black')
btn_generate.pack(pady=5)

a.mainloop()
```
