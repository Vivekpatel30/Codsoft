```python
from tkinter import *
def bClick(number):
    global val
    val=val+str(number)
    data.set(val)
def bClear():
    global val
    val=''
    data.set("")
def bEquals():
    global val
    result=str(eval(val))
    data.set(result)

a=Tk()

a.title("my calculator")
a.geometry("360x450")
val=''
data=StringVar()
disp=Entry(a,bd=29  ,bg='black',fg='white',textvariable=data,justify='right',font=('white',20))
disp.grid(row=0,columnspan=6)
b9=Button(a,text='9',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(9))
b9.grid(row=1,column=0)
b8=Button(a,text='8',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(8))
b8.grid(row=1,column=1)
b7=Button(a,text='7',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(7))
b7.grid(row=1,column=2)
b_add=Button(a,text='+',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick('+'))
b_add.grid(row=1,column=3)
b6=Button(a,text='6',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(6))
b6.grid(row=2,column=0)
b5=Button(a,text='5',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(5))
b5.grid(row=2,column=1)
b4=Button(a,text='4',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(4))
b4.grid(row=2,column=2)
b_sub=Button(a,text='-',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick('-'))
b_sub.grid(row=2,column=3)
b3=Button(a,text='3',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(3))
b3.grid(row=3,column=0)
b2=Button(a,text='2',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(2))
b2.grid(row=3,column=1)
b1=Button(a,text='1',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(1))
b1.grid(row=3,column=2)
b_multi=Button(a,text='*',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick('*'))
b_multi.grid(row=3,column=3)
b_c=Button(a,text='C',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=bClear)
b_c.grid(row=4,column=0)
b_0=Button(a,text='0',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick(0))
b_0.grid(row=4,column=1)
b_eql=Button(a,text='.',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick('.'))
b_eql.grid(row=4,column=2)
b_div=Button(a,text='รท',bg='pink',fg="black",font=('blue',12,'bold'),bd=12,height=2,width=6,command=lambda:bClick('/'))
b_div.grid(row=4,column=3)
b_00=Button(a,text='00',bg='maroon',fg="yellow",font=('blue',12,'bold'),bd=12,height=2,width=15,command=lambda:bClick('00'))
b_00.grid(row=5,column=0,columnspan=2)
b_dot=Button(a,text='=',bg='maroon',fg="yellow",font=('blue',12,'bold'),bd=12,height=2,width=15,command=bEquals)
b_dot.grid(row=5,column=2,columnspan=2)

a.mainloop() 


    
```
