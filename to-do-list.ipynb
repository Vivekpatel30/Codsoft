```python
from tkinter import *


a= Tk()
a.title("To-Do List")
a.config(bg='black')



def add_task():
  task = entry_text.get()
  if task:
    listbox.insert(END, task)
    entry_text.delete(0, END)

    
def delete_task():
  selected = listbox.curselection()
  if selected:
    listbox.delete(selected[0])


entry_text = Entry(a, 
                   width=28,
                   font=('arial',20,'bold'),
                   bg='pink'
                  )
entry_text.pack()


add_button = Button(a,
                    text="Add",
                    bg='yellow',
                    height=2,
                    width=7,
                    font=('arial',10,'bold'),
                    command=add_task
                   )
add_button.pack()


listbox = Listbox(a,
                  height=7,
                  width=28,
                  font=('arial',20,'bold'),
                  bg='pink',
                  selectbackground='gray'
                 )
listbox.pack()


delete_button = Button(a, 
                       text="Delete",
                       height=2,
                       width=7,
                       font=('arial',10,'bold'),
                       bg='yellow', 
                       command=delete_task)
delete_button.pack()

a.mainloop()
```
